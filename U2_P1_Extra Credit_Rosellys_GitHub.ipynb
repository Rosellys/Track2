# Rock Paper Scissors
## Game with Input for Selections

```python
# Create Variables
Player = 0
Computer = 0
Rosellys_Score = 0
Computer_Score = 0
Score = 0
Options = ['Rock','Paper','Scissors']
Computer_List =[]
Message_List = []
Score_List = []
R_Score_List = []
C_Score_List = []
Rosellys_List = []
```

```python
# Game While/ Inputs
Play = True
while Play == True :
    Computer = Options[randint(0,2)]
    Computer_List.append(Computer)
    Rosellys = input('Rock, Paper, Scissors... SHOOT!')
    Rosellys_List.append(Rosellys)
    if Rosellys == Computer :
        Message ='Stalemate!'
    
    elif Rosellys == 'Rock' :  
        if Computer == 'Scissors' :
                Message = 'Rosellys wins! || ' + Rosellys +  ' obliterates '+ Computer
                Rosellys_Score = Rosellys_Score + 1
        else:
                Message = 'Computer wins! || ' + Computer + ' engulfs ' + Rosellys
                Computer_Score = Computer_Score + 1
    elif Rosellys == 'Scissors' :     
        if Computer == 'Paper' :
                Message = 'Rosellys wins! || '+  Rosellys +  ' shreds ' + Computer
                Rosellys_Score = Rosellys_Score + 1
        else:
                Message = 'Computer wins! || ' + Computer + ' obliterates ' + Rosellys
                Computer_Score = Computer_Score + 1
    elif Rosellys == 'Paper' :     
        if Computer == 'Rock' :
                Message = 'Rosellys wins! || ' + Rosellys + ' engulfs ' + Computer
                Rosellys_Score = Rosellys_Score + 1
        else:
                Message = 'Computer wins! || ' + Computer + ' shreds ' + Rosellys
                Computer_Score = Computer_Score + 1
    else :
        print('Can you not spell?',  'Try again and spell it correctly this time.')
        
    Score = 'Rosellys: ' + str(Rosellys_Score) + ' || Computer: ' + str(Computer_Score)
    Score_List.append(Score)
    R_Score_List.append(Rosellys_Score)
    C_Score_List.append(Computer_Score)
    Message_List.append(Message)
    Again = input ('Would you like to dule again? (Yes/No)')
    if Again == 'Yes' :
        Play = True
    else :
        Play = False
  
```

```python
# Dictionary of lists  
dict = {'Rosellys': Rosellys_List,  'Computer': Computer_List, 
        'Message' : Message_List,   'Score': Score_List, 
       }  
    
df = pd.DataFrame(dict) 
    
df  
```

```python
# Final Ruling
if Rosellys_Score == Computer_Score :
    print('Even Stevens; lets play another 5 rounds')
elif Rosellys_Score > Computer_Score :
    print('Rosellys is VICTORIOUS!.!.!.!  ' + Score)
else :
    print('Rosellys is such a noob. Who lets the computer beat them? ...' + Score)
```
